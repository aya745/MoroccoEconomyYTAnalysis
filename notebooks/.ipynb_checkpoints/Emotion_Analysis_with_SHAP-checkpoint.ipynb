```python
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import torch
import transformers
import datasets
import shap
```

```python
data = pd.read_csv('C:\\Users\\tech\\Desktop\\MoroccoEconomyYTAnalysis\\notebooks\\cleaned_transdataset.csv', encoding='utf-8')
```

```python
data.head()
```

```python
# load the model and tokenizer
tokenizer = transformers.AutoTokenizer.from_pretrained("nateraw/bert-base-uncased-emotion", use_fast=True)
#tokenizer = transformers.AutoTokenizer.from_pretrained("nlptown/bert-base-multilingual-uncased-sentiment", use_fast=True)
#tokenizer = transformers.AlbertForTokenClassification.from_pretrained("cmarkea/distilcamembert-base-sentiment", use_fast=False)

model = transformers.AutoModelForSequenceClassification.from_pretrained("nateraw/bert-base-uncased-emotion").cpu()
#model = transformers.AutoModelForSequenceClassification.from_pretrained("nlptown/bert-base-multilingual-uncased-sentiment").cpu()
#model = transformers.AutoModelForSequenceClassification.from_pretrained("cmarkea/distilcamembert-base-sentiment").cpu()
```

```python
# build a pipeline object to do predictions
pred = transformers.pipeline("text-classification", model=model, tokenizer=tokenizer, device=-1, return_all_scores=True)
```

```python
explainer = shap.Explainer(pred)

str(data['divided_captions'][0])[:512]

shap_values = explainer(str(data['divided_captions'][0])[:12])

```

```python
shap_values
```

```python
shap.plots.text(shap_values)
```

```python
# Assuming shap_values has shape (num_samples, num_features, num_categories)
# Access the "anger" category for the first sample
shap.plots.text(shap_values[0, :, "anger"])
```

```python
# Assuming shap_values has shape (num_samples, num_features, num_categories)
# Access the "joy" category for the first sample
shap.plots.text(shap_values[0, :, "joy"])
```

```python
logit_explainer = shap.Explainer(shap.models.TransformersPipeline(pred, rescale_to_logits=True))
logit_shap_values = logit_explainer(data['divided_captions'][:3])
shap.plots.text(logit_shap_values)
```
